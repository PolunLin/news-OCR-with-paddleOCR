```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
from paddleocr import PaddleOCR,draw_ocr
# Paddleocr supports Chinese, English, French, German, Korean and Japanese.
# You can set the parameter `lang` as `ch`, `en`, `fr`, `german`, `korean`, `japan`
# to switch the language model in order.
font = ImageFont.truetype('msjhbd.ttc')
ocr = PaddleOCR(use_angle_cls=False, 
                lang='chinese_cht', 
                drop_score=0.3,
                save_crop_res=True,
                det_db_score_mode	='slow',
                det_limit_type='min',
                det_limit_side_len=2000,
                det_db_box_thresh =0.1,
                det_db_thresh=0.2) # need to run only once to download and load model into memory

# # draw result
# from PIL import Image
def image_raw_text(img_path,OCR_result,save_path):
    '''
    show original image and put the raw text next to it. 

    input:
        img_path:original image path
        save_path:save path of this out image 
        OCR_result: your result of paddleOCR
    oupput:
        None:only save image in the path
    '''
    image = Image.open(img_path).convert('RGB')
    boxes = [line[0] for line in OCR_result]
    txts = [line[1][0] for line in OCR_result]
    scores = [line[1][1] for line in OCR_result]
    im_show = draw_ocr(image, boxes, txts, scores, font_path='msjhbd.ttc')
    im_show = Image.fromarray(im_show)
    im_show.save(save_path)


##
def image_with_text(image_path, OCR_result, save_path,font,color='yellow',width=2,only_text=False):
    '''
    show original image and put the raw text on it. 

    input:
        img_path:original image path
        save_path:save path of this out image 
        OCR_result: your result of paddleOCR
        font:the font of the text on the image
        color: the color of the lines
        width:the width of the lines
        only_text: only text or not
    oupput:
        None:only save image in the path
    '''
    image = Image.open(image_path)
    draw = ImageDraw.Draw(image)
    if only_text:
        image = Image.new("RGB", (2000, 2000), (255, 255, 255))
    else:
        for bound in OCR_result:
            p0, p1, p2, p3 = bound[0]
            draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
            draw.text([*p0, *p1, *p2, *p3, *p0],bound[1][0], font=font, fill="#000")
    image.save(save_path)

```

```python
img_path = 'data/example_news.jpg'
result = ocr.ocr(img_path, cls=False)
for line in result:
    print(line)

```

```python
image_raw_text(img_path,result,"data/output/image_raw_text.jpg")
image_with_text(img_path, result, "data/output/image_with_text.jpg",font,color='yellow',width=2,only_text=False)
image_with_text(img_path, result, "data/output/only_text_text.jpg",font,color='yellow',width=2,only_text=False)
```
